## Data Loading and Initial Inspection

This notebook loads the raw dataset, performs initial checks, and creates a working sample for further analysis.

```python
import pandas as pd
import numpy as np
from pathlib import Path
```

### Project Paths Setup

Defining paths for raw data, processed data, and outputs.

```python
# Move from notebooks/ → project root
PROJECT_ROOT = Path.cwd().resolve().parent

DATA_RAW = PROJECT_ROOT / "data" / "raw"
DATA_PROCESSED = PROJECT_ROOT / "data" / "processed"
OUTPUTS = PROJECT_ROOT / "outputs"

print("Project root:", PROJECT_ROOT)
print("Raw data path:", DATA_RAW)
```

### Load Raw Dataset

Loading selected columns from the raw dataset for analysis.

```python
file_path = DATA_RAW / "accidents.csv"

print("File exists:", file_path.exists())
print("File path:", file_path)
```

### Initial Data Inspection

Checking structure, data types, missing values, and severity distribution.

```python
cols_to_use = [
    "Severity",
    "Start_Time",
    "Start_Lat",
    "Start_Lng",
    "Weather_Condition",
    "Visibility(mi)",
    "Junction",
    "Traffic_Signal"
]

df = pd.read_csv(
    file_path,
    usecols=cols_to_use,
    low_memory=False
)

print("Shape:", df.shape)
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df["Severity"].value_counts(dropna=False).sort_index()
```

### Create Sample Dataset

Creating a smaller sample dataset for faster processing and experimentation.

```python
sample_df = df.sample(n=300000, random_state=42)

print("Sample shape:", sample_df.shape)
```

### Save Sample Dataset

Saving the sampled dataset for use in preprocessing and analysis.

```python
DATA_PROCESSED.mkdir(parents=True, exist_ok=True)

sample_path = DATA_PROCESSED / "accidents_sample_300k.csv"

sample_df.to_csv(sample_path, index=False)

print("Saved to:", sample_path)
```
